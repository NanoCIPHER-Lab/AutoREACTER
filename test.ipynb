```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw

# Define the reaction: Ethene + Benzene -> Ethylbenzene
rxn_smarts = "[C:1]=[C:2].[c:3]1[c:4][c:5][c:6][c:7][c:8]1>>[c:3]1[c:4][c:5][c:6][c:7][c:8]1[C:1][C:2]"
rxn = AllChem.ReactionFromSmarts(rxn_smarts)

# Visualize it
img = Draw.ReactionToImage(rxn)
img.show()
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import Image # Necessary to show the binary data as an image

# 1. Define your molecules
reac1 = 'CC(=O)O' 
reac2 = 'OCC'    
prod = 'CCOC(C)=O' 
reac1 = Chem.CombineMols(Chem.MolFromSmiles(reac1), Chem.MolFromSmiles(reac2))
# 2. Create the reaction object
full_rxn_smarts = f"{reac1}.{reac2}>>{prod}"
rxn = AllChem.ReactionFromSmarts(full_rxn_smarts, useSmiles=True)

# 3. Draw with high quality
d2d = rdMolDraw2D.MolDraw2DCairo(800, 300)
d2d.DrawReaction(rxn)
d2d.FinishDrawing()

# 4. Grab the binary data and display it
png_data = d2d.GetDrawingText()
Image(png_data)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw

# Define a list of reaction SMILES (Reactants >> Products)
reaction_list = [
    "CC(=O)O.OCC>>CCOC(C)=O",              # Esterification
    "C=C.ClCl>>FCC(Cl)Cl",                 # Halogenation (simplified)
    "c1ccccc1.BrBr>>c1ccccc1Br.Br",        # Bromination
    "CC(=O)C.O>>CC(O)(O)C"                 # Hydration
]

# Convert strings into RDKit Reaction objects
rxns = [AllChem.ReactionFromSmarts(r, useSmiles=True) for r in reaction_list]

# Create a grid image
img = Draw.ReactionsToGridImage(rxns, reactionsPerRow=2, subImgSize=(400, 200))
display(img)
```

```python
# Note the middle section between the two '>' signs
complex_rxn_string = "CC(=O)O.OCC > CCOC(C)=O"

rxn = AllChem.ReactionFromSmarts(complex_rxn_string, useSmiles=True)
Draw.ReactionToImage(rxn)
```

```python
def add_sprinkles(function):
    def wrapper(*args, **kwargs):
        print("Adding sprinkles on top!")
        function(*args, **kwargs)
    return wrapper

def add_fudge(function):
    def wrapper(*args, **kwargs):
        print("Adding fudge on top!")
        function(*args, **kwargs)
    return wrapper

@add_sprinkles
@add_fudge
def get_ice_cream(flavor):
    print(f"Here is your {flavor} ice cream.")

def get_ice_cream_2():
    print("Here is your Ice cream.")

get_ice_cream_2()
get_ice_cream("Chocolate")
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.is_alive = True

    def __repr__(self):
        return f"(name={self.name}, age={self.age}, is_alive={self.is_alive})"

    def __eq__(self, other):
        if not isinstance(other, Person):
            return NotImplemented
        return self.name == other.name and self.age == other.age

person1 = Person("Spongebob", 30)
person2 = Person("Patrick", 35)
person3 = Person("Patrick", 35)

print(person1)
print(person2)
print(person3)

print(person1 == person2)  # False
print(person2 == person3)  # True
```
